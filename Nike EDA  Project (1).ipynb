3. Equipment and Accessories

```python
# Top 10 state by sum of Total sales and Profit

y=df.groupby("State")["Total Sales"].sum().reset_index()
y=y.sort_values(by="Total Sales",ascending=False)
y=y.head(10)
y
```

```python
plt.figure(figsize=(15,5))
plt.title("Top 10 States by total sales");
sns.barplot(data=y,x="State",y="Total Sales",palette="rainbow")
plt.show()
```

```python
# sum of profit by each region.


x=df.groupby("Region")["Operating Profit"].sum().reset_index()

sns.set_style("darkgrid")
plt.title("Percentage of profits by each region")
plt.pie(x["Operating Profit"],labels=x["Region"],autopct="%0.2f%%");
plt.pie([1],radius=0.4,colors="w");
```

```python
# line chart to show total sales and profit trends by invoice date 

sns.set_style("darkgrid")
plt.figure(figsize=(15,5))
plt.title("Total Sales by invoice date")
sns.lineplot(data=df,x="Invoice Date",y="Total Sales",color="crimson");



sns.set_style("darkgrid")
plt.figure(figsize=(15,5))
plt.title("Profit Trends by invoice date")
sns.lineplot(data=df,x="Invoice Date",y="Operating Profit",color="teal");
```

```python
# pie chart to show maximum unit sold by each product

s=df.groupby("Product")["Units Sold"].max().reset_index()

plt.pie(s["Units Sold"],labels=s["Product"],autopct="%0.1f%%");
```

```python
# average total sales by each retailer

r= df.groupby("Retailer")["Total Sales"].mean().reset_index()

plt.figure(figsize=(13,5))
plt.title("Average Total Sales By Retailer");
sns.barplot(data=r,y="Retailer",x="Total Sales",palette="plasma");
```

```python
# Price Per Unit According to the product

plt.figure(figsize=(10,5))
plt.title("Price per Unit according to Product")
sns.histplot(data=df,x="Price per Unit",kde=True,hue="Product");
```

```python
# Total sales by each city and retailer

plt.figure(figsize=(20,5))
plt.title("Total sales by each city and retailor");
sns.barplot(data=df,x="City",y="Total Sales",palette="Set2",hue="Retailer");
plt.xticks(rotation=75);
```

📌 Final Insights & Conclusion – Nike EDA Project
1. Sales Trends

The sales analysis shows clear patterns, with certain time periods/seasons driving higher demand.

Some products/categories outperform consistently, while others have lower traction.

Sales are concentrated in a few key markets, which can guide inventory planning.

2. City-Wise Performance

From the charts, it is clear that some cities dominate total sales, highlighting Nike’s strongest markets.

Other cities show moderate or low performance, which indicates potential growth opportunities if marketing and supply chain efforts are optimized there.

3. Retailer Comparison

Retailers contribute differently to sales volume. A few retailers are clear leaders, while others lag behind.

The differences across retailers suggest that Nike can strengthen partnerships with top performers and work on improving efficiency with weaker ones.

4. ETA (Delivery Time) Analysis

The average ETA varies across cities and retailers, showing inconsistencies in delivery performance.

Some retailers achieve faster delivery times, which directly improves customer satisfaction.

High ETA in certain regions suggests the need for logistics optimization or alternative supply chain strategies.

5. Customer Experience Implications

Customers in cities with faster deliveries are more likely to stay loyal and repeat purchases.

Long delays may hurt customer trust, making ETA a critical metric for Nike’s future strategy.

Retailers with better delivery performance can act as benchmarks for improvement.
